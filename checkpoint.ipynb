```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from scipy.cluster import hierarchy
from scipy.spatial import distance_matrix
```

### Let's get familiar with our dataset

```python
#Import the csv file
data= pd.read_csv('CC GENERAL.csv', header = 'infer')
```

```python
#the column min payments is full of nan values, so we need to drop it
data.drop(['MINIMUM_PAYMENTS'], axis=1, inplace = True)
data.fillna(data.CREDIT_LIMIT.mean(), inplace=True)
data
```

### Preprocessing time

```python
#remove the object column:
data.drop(['CUST_ID'], axis = 1, inplace=True)
scaled_data = StandardScaler().fit_transform(data)
scaled_data_df = pd.DataFrame(scaled_data)
scaled_data_df
```

```python
# We gonna use the pca method with 2 components in order to simplify the taks after that
pca = PCA(n_components=2).fit(scaled_data)
x_pca = pca.transform(scaled_data)
x_pca.shape
```

```python
scores_df = pd.DataFrame(x_pca, columns=['PC1', 'PC2'])
scores_df
```

```python
plt.figure(figsize=(20,15))
plt.scatter(x_pca[:,0], x_pca[:,1], alpha=0.9)
```

### High time to cluster our data

```python
myLinkage =  hierarchy.linkage(distance_matrix(x_pca,x_pca),'complete')   
#Dendogram 
dendro = hierarchy.dendrogram(myLinkage)
```

```python
labels = hierarchy.fcluster(myLinkage,500, criterion='distance')                      #  Cut at the height (distance) = 0.23 <= change this value at will.
pd.Series(labels).value_counts()
```

```python
plt.figure(figsize=(20,15))
plt.scatter(x_pca[:,0], x_pca[:,1], c= labels, alpha=0.7)
```

```python
agglomerative = AgglomerativeClustering(n_clusters=3)              
agglomerative = agglomerative.fit(x_pca)                                   
#Scatter plot to vizualise 
plt.figure(figsize=(20,15))
plt.scatter(x_pca[:,0], x_pca[:,1], c= agglomerative.labels_, alpha=0.7) 
```

### Let's work with Kmeans now

```python
#now onto kmeans
km = KMeans(n_clusters=3)
y_km = km.fit_predict(x_pca)
plt.figure(figsize=(20,15))
plt.scatter(x_pca[y_km == 0,0], x_pca[y_km == 0,1], c= 'darkblue', alpha=0.7) 
plt.scatter(x_pca[y_km == 1,0], x_pca[y_km == 1,1], c= 'pink', alpha=0.7) 
plt.scatter(x_pca[y_km == 2,0], x_pca[y_km == 2,1], c= 'purple', alpha=0.7) 
```

```python
#now onto kmeans
km = KMeans(n_clusters=4)
y_km = km.fit_predict(x_pca)
plt.figure(figsize=(20,15))
plt.scatter(x_pca[y_km == 0,0], x_pca[y_km == 0,1], c= 'darkblue', alpha=0.7) 
plt.scatter(x_pca[y_km == 1,0], x_pca[y_km == 1,1], c= 'pink', alpha=0.7) 
plt.scatter(x_pca[y_km == 2,0], x_pca[y_km == 2,1], c= 'purple', alpha=0.7) 
```

```python
#now onto kmeans
km = KMeans(n_clusters=2)
y_km = km.fit_predict(x_pca)
plt.figure(figsize=(20,15))
plt.scatter(x_pca[y_km == 0,0], x_pca[y_km == 0,1], c= 'darkblue', alpha=0.7) 
plt.scatter(x_pca[y_km == 1,0], x_pca[y_km == 1,1], c= 'pink', alpha=0.7) 
plt.scatter(x_pca[y_km == 2,0], x_pca[y_km == 2,1], c= 'purple', alpha=0.7) 
```

```python
# Let's find the optimal k value using elbow method
Y=[]
for i in range(1,10):
        Y.append(KMeans(n_clusters=i).fit(x_pca).inertia_)
plt.xlabel("Number of clusters")
plt.ylabel("TSS")
plt.plot(range(1,10), Y)
```

the best K value is here 3

Conclusion:
we can clearly see from the plots that each clustering method gives a different result, the simple hierarchichal methods automatically give clusters, when plotting the dendogram, it tok really long time(8mins), while kmeans was just quick with less than 2 seconds with high performance and good clustering based on the k value

```python
print(f"agglomerative score: {silhouette_score(x_pca, agglomerative.fit_predict(x_pca))}")
print(f"hierarchical score: {silhouette_score(x_pca, labels)}")
print(f"kmeans score: {silhouette_score(x_pca, KMeans(n_clusters=3).fit_predict(x_pca))}")
```
